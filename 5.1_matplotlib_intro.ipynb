```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, np.pi*4, num=200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
z = np.cos(x)

plt.plot(x,y,'r', label='sin(x)')
plt.plot(x,z, 'b', label = 'cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,6])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y,label='sin(x)')
axes.legend()

axes.set_xlabel()
axes.set_title()

plt.show()
```
