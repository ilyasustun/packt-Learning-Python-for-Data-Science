```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
col_names = ['mpg', 'cylinders', 'dis', 'hp', 'weight', 'acc', 'year', 'origin', 'name']
```

```python
df = pd.read_csv('auto-mpg.data', sep='\s+', names=col_names)
df
```

```python
plt.hist(df['mpg'])
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.hist(df['mpg'], bins=20)
plt.xlabel('mpg')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x=df['weight'], y=df['mpg'], alpha=0.5)

plt.ylabel('mpg')
plt.xlabel('weight')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.hist2d(x=df['weight'], y=df['mpg'])

plt.ylabel('mpg')
plt.xlabel('weight')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.hexbin(x=df['weight'], y=df['mpg'], gridsize=(35,15), cmap='coolwarm')

plt.ylabel('mpg')
plt.xlabel('weight')
plt.colorbar()
plt.show()
```
